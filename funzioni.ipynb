```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generare un array Id di valori equispaziati in un intervallo
np.linspace(0.,5.,11)  #11 valori equispaziati tra 0 e 5 
```

Numpy implementa tutte le principali funzioni matematiche.
Per esempio la funzione $e^x$ Ã¨ implementata da `np.exp(x)`.
La gaussiana semplice ha espressione $e^(-x^2)$.

```python
def gaussianaSemplice(x):
    return np.exp(-x**2)
    
x=np.linspace(-10.,10.,1000) #ascissa
y=gaussianaSemplice(x) #broadcast

plt.plot(x,y)
```

24/10725

```python
# fig Ã¨ la tela su cui faccio i grafici
# ax sono gli assi di un grafico posto sulla figure
fig, ax=plt.subplots(figsize=(5,2))
ax.plot(x,y,color="#EDDDD4",linewidth=2, label="$f(x)=e^(-x^2)$")
ax.grid()
ax.legend()
fig.set_facecolor("#C44536")
ax.set_facecolor("#197278")
ax.set_title("Il primo grafico")
ax.set_xlabel("x")
ax.set_ylabel("y")

plt.show()
```

```python
f=lambda x:np.sin(x)/x
g=lambda x:np.cos(x)
```

```python
x=np.linspace(-4*np.pi, 4*np.pi, 400)
fig,ax=plt.subplots(figsize=(10, 5))

ax.plot(x,f(x), color="#EDDDD4", linewidth=2, label=r"$f(x)=\frac{\sin(x)}{x}$")
ax.plot(x,g(x), color="#83C5BE", linewidth=2, label=r"$g(x)=\cos(x)$")

ax.grid()
ax.legend()
fig.set_facecolor("#C44536")
ax.set_facecolor("#197278")
ax.set_title("Il primo grafico")
ax.set_xlabel("x")
ax.set_ylabel("y")

plt.show()
```
