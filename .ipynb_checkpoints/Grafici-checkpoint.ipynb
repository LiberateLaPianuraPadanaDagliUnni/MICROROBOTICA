```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def estimatePI(n_lanci):
    raggio = 1.0
    x = np.random.uniform(-raggio, raggio, n_lanci)
    y = np.random.uniform(-raggio, raggio, n_lanci)
    dentro_cerchio = x**2 + y**2 < raggio**2
    n_dentro = np.count_nonzero(dentro_cerchio)
    return 4 * n_dentro / n_lanci
```

```python
arrayNlanci = np.array([10**n for n in range(2, 8)])
stimePI = np.array([estimatePI(n) for n in arrayNlanci])
erroreAssoluto = np.abs(stimePI - np.pi)
```

```python
fig, ax = plt.subplots(figsize=(5, 2))
ax.plot(arrayNlanci, stimePI)
ax.set_xlabel("Numero di lanci")
ax.set_ylabel("Stime di $\pi$")
ax.grid()
ax.hlines(np.pi,0,10_000_000, colors="red", linestyles="dotted")
ax.set_ylim([3.1,3.2])
plt.show()
```
